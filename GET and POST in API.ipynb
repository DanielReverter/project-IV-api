## Get with params

```python
params = {"language":"ca"}
```

```python
date = "2020-03-19"
```

```python
response = requests.get(f"http://127.0.0.1:5000/translate/{date}?language={params['language']}").json()
```

# POST

### Create new tweet

```python
data = {"date":"2022-08-15", "tweet":"Tesla is great :)", "language":"en"}
```

```python
url = "http://127.0.0.1:5000/newtweet/"
```

```python
requests.post(url, data=data)
```
